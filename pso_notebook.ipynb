# Particle Swarm Optimization

```python
# imports
import gym

from pso import Particle
from pso import particle_swarm
```

```python
# make the environment and observe the state and action spaces
env = gym.make('CartPole-v1')
print('observation space:', env.observation_space)
print('action space:', env.action_space)
```

```python
# initiate the optimization
y = particle_swarm(25,.5,1,2,0.25)
```

```python
# observe a trained particle
state = env.reset()
for t in range(1000):
    env.render()
    action = y.act(state, y.bw)
    state, reward, done, _ = env.step(action)
    if done:
        env.close()
        break
```
